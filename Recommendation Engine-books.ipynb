```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('book.csv',encoding='cp1252')
```

```python
df.head()
```

```python
df.shape
```

```python
df.iloc[:,1].duplicated()
```

```python
df = df.rename(columns = {'User.ID': 'UserID', 'Book.Title': 'BookTitle','Book.Rating': 'BookRating'})
df
```

```python
#number of unique users in the dataset
len(df.UserID.unique())
```

```python
df.BookTitle.unique()
```

```python
book_df = df.pivot_table(index='UserID',columns='BookTitle',values='BookRating')
book_df
```

```python
book_df
```

```python
#Impute those NaNs with 0 values
book_df.fillna(0, inplace=True)
```

```python
book_df
```

```python
#Calculating Cosine Similarity between Users
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
user_sim = 1 - pairwise_distances( book_df.values,metric='cosine')
```

```python
user_sim
```

```python
user_sim.shape
```

```python
#Store the results in a dataframe
user_sim_df = pd.DataFrame(user_sim)
user_sim_df 
```

```python
#Set the index and column names to user ids 
user_sim_df.index = df.UserID.unique()
user_sim_df.columns = df.UserID.unique()
```

```python
user_sim_df
```

```python
user_sim_df.iloc[0:5, 0:5]
```

```python
np.fill_diagonal(user_sim, 0)
user_sim
```

```python
#Most Similar Users
user_sim_df.idxmax(axis=1)
```
